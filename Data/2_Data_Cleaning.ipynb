```python
## Handling missing value in the close_date column
## Missing close_date values were kept as they represent active listings that have not been closed yet
df_listings['close_date'] = pd.to_datetime(
    df_listings['close_date'],
    errors='coerce'
)
```

Data Cleaning : Customers Monthly Metrics

```python
import pandas as pd
df_cmm = pd.read_csv(r'E:\5_Portofolio Data Analyst\1_Real_Estate_Analysis\Real_Estate_Data.csv\customer_monthly_metrics.csv')
df_cmm.head()
```

```python
df_cmm.info()
```

```python
df_cmm['month'] = pd.to_datetime(df_cmm['month'])
df_cmm['signup_month'] = pd.to_datetime(df_cmm['signup_month'])
df_cmm['churn_month'] = pd.to_datetime(df_cmm['churn_month'])
```

```python
df_cmm.duplicated().sum()
```

```python
df_cmm.isnull().sum()
```

```python
df_cmm['churned'] = df_cmm['churned'].fillna('UNKNOWN')
```

```python
#Added a churn_status column to identify customers with missing churn_month values as active customers
df_cmm['churn_status'] = df_cmm['churn_month'].apply(
    lambda x: 'Churned' if pd.notna(x) else 'Active'
)
```
